```python
import pymysql
import time
```

```python
db_name = 'test_wos_cut_full'

client_config = {'unix_socket':'/project2/jevans/study_dbs/mysql/.sql.sock',
                'database': db_name}
```

```python
# prototyping -
#  count X & count Y

# note - double-counting some papers that cite both A and members of X

start_time = time.time()

year = 1978  # for the year specified here
CUTOFF = 5 # threshold on the ratio

db = pymysql.connect(**client_config)

cursor = db.cursor()
sql = '''
        SELECT TX.P0, TX.countX, TY.countY, TY.countY/TX.countX as ratio
        FROM
        
            (
            SELECT P0, count(P0) as countY
            FROM
                (
                SELECT P0, P1, P2, count(P2) as count
                FROM
                    (
                    SELECT P0, P1, Refs2.wos_id as P2
                    FROM
                        (
                        SELECT Pubs.wos_id as P0, Refs.wos_id as P1
                        FROM publications as Pubs
                        JOIN the_references as Refs
                        ON Pubs.wos_id = Refs.uid
                        WHERE Pubs.pubyear = {}
                        ) AS T1
                    JOIN the_references as Refs2
                    ON P1 = Refs2.uid
                    ) AS T2
                GROUP BY P0, P2
                ORDER BY count DESC
                ) AS T3
            GROUP BY P0
            ORDER BY countY DESC
            ) as TY
            
        JOIN
        
            (SELECT P0, count(*) as countX
            FROM
                    (
                    SELECT Pubs.wos_id as P0, Refs.wos_id as P1
                    FROM publications as Pubs
                    JOIN the_references as Refs
                    ON Pubs.wos_id = Refs.uid
                    WHERE Pubs.pubyear = {}
                    ) AS T1
            GROUP BY P0
            ORDER BY countX DESC) AS TX
        
        ON TY.P0 = TX.P0
        HAVING ratio > 5
        
        '''.format(year, year)

cursor.execute(sql)

results = []
for i,result in enumerate(cursor):
    A = result[0]
    sizeX = result[1]
    sizeY = result[2]
    ratio = float(result[3])
    results.append( {'A': A,
                    'sizeX': sizeX,
                    'sizeY': sizeY,
                    'ratio': ratio} )

cursor.close()
db.close()

end_time = time.time()
elapsed = end_time - start_time
print("elapsed (s): {}".format(elapsed))
```

```python
import json

save_obj = {'elapsed': elapsed,
            'results': results,
            'N_results': len(results)}

with open('query3_results.json','w') as f:
    json.dump(save_obj, f)
```
